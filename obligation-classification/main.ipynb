```python
a = "The Parties have jointly drafted the contract guide set out in Schedule 19 (Contract Guide) and both Parties will apply this guide in its training of team members and as a support tool in the day to day management of the relationship."
a = "Notwithstanding Paragraph 12.2 Of Schedule 13.1 (Rail Industry Initiatives) By No Later Than 31 December 2018 And Throughout The Remainder Of The Franchise Period Provide The Mobile Communication Services In Accordance With Schedule 13.1 (Rail Industry Initiatives) In Both First Class And Standard Class Accommodation On All Of The Train Fleet Used For The Provision Of Passenger Services Such Mobile Communication Services To Provide The Following Levels As A Minimum"                                                                                                                                                                       
# a = "The Franchisee Shall Working With One Or More Appropriate Telecommunication Service Providers Or Mobile Network Operators Use All Reasonable Endeavours To Procure That 100 Coverage Of The Network Over Which The Passenger Services Operate For Both Voice And Data Communications Is Achieved By No Later Than 31 December 2018 In Order To Support The Provision Of The Mobile Communication Services. "
a = "Both the supplier and the franchise shall meet quarterly"

# responsibility of bot, examples from tracker
{ "DESCRIPTION" : "The Project Brief will be prepared and signed off by both the customer and Telefónica and then kept up to date throughout the full project execution by the Global Project Manager." }



parsedData = parser(a)
for span in parsedData.sents:
    sent = [parsedData[i] for i in range(span.start, span.end)]
    for token in sent:
        print(token.orth_, token.pos_)
```

```python
# Let's look at the sentences
sents = []
# the "sents" property returns spans
# spans have indices into the original string
# where each index value represents a token
for span in parsedData.sents:
    # go from the start to the end of each span, returning each token in the sentence
    # combine each token using join()
    sent = ''.join(parsedData[i].string for i in range(span.start, span.end)).strip()
    sents.append(sent)

for sentence in sents:
    print(sentence)
```

```python
# Let's look at the part of speech tags of the first sentence
for span in parsedData.sents:
    sent = [parsedData[i] for i in range(span.start, span.end)]
    for token in sent:
        print(token.orth_, token.pos_)

```

```python
import pymongo
from pymongo import MongoClient
import numpy as np
import pandas as pd


client = MongoClient()
db = client.obligations
collection = db.obligation_tracker
c = collection.find({ 'DESCRIPTION': { '$exists': True}})
obligations = []
responsbility = -1



for a in c:
    if a['RESPONSIBILITY'] in ["Supplier", "Customer" ]:
        obligations += [{'description': a['DESCRIPTION'], 'responsibility': 0, 'supplier':1 }]
    elif a['RESPONSIBILITY'] in ["Bloomberg", "Client", "UCLH", "Vodafone"]:
        obligations += [{'description': a['DESCRIPTION'], 'responsibility': 1, 'supplier':0 }]
    elif a['RESPONSIBILITY'] in ["Both - Bloomberg & Supplier", "Both - Client & Supplier", "Both - UCLH & Supplier", "Both - UCLH & Customer", "Both - Vodafone & Supplier"]:
        obligations += [{'description': a['DESCRIPTION'], 'responsibility': 2, 'supplier':0 }]
    else:
        print(a['RESPONSIBILITY'])
    
len(obligations)
df = pd.DataFrame(obligations)
```
